```
#Prediction
from imageai.Prediction.Custom import CustomImagePrediction
#create an instance/object of CustomImagePrediction
prediction = CustomImagePrediction()
```

```
#set the model type
prediction.setModelTypeAsResNet()
#set the model path
prediction.setModelPath("idenprof_061-0.7933.h5")
#set the json path
prediction.setJsonPath("idenprof_model_class.json")
```

```
#specify number of objects
prediction.loadModel(num_objects=10)
```

```
#prediction
predictions, probabilities = prediction.predictImage("download.jpg", result_count=3)
```

```
#combine prediction probabilities using zip()
```

```
'''Zip
use to map the indexes of containers.
'''
name = ['John','Alex','Smith']
rollNo = [1,2,3]
mapped = zip(name,rollNo)
print(list(mapped))
```

```
for eachPrediction, eachProbability in zip(predictions,probabilities):
  print(eachPrediction,':',eachProbability)
```
